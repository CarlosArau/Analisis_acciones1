### 1. Apple 

**Paso 1:  Validar código HTML**

```python
apple_html = pd.read_table(r"C:\Users\Carla\Downloads\2. Acciones_tecnologicas\apple.txt")
print(apple_html)
```

**Paso 2: Transformar codigo html a DataFrame**

```python
apple = pd.read_html(r"C:\Users\Carla\Downloads\2. Acciones_tecnologicas\apple.txt")[0]
apple
```

**Paso 3: Renombrar columnas**

```python
apple.rename(columns={"Close Close price adjusted for splits.":"Close"}, inplace = True)
apple.head()
```

```python
apple_data = apple[["Date","Open" , "High" , "Low" , "Close" , "Volume"]]
apple_data.head()
```

**Paso 4 : Ver el tipo de datos de cada columna**

```python
print(apple_data.dtypes)
```

**Paso 5 : Modificar los tipos objeto a float (decimales)**

```python
columnas = ['Open', 'High', 'Low', 'Close', 'Volume']

apple_data = apple_data.copy()
apple_data[columnas] = apple_data[columnas].replace(r'[^\d.]', '', regex=True)
apple_data[columnas] = apple_data[columnas].astype(float)
apple_data.info()
```

**Paso 6: Cambiar formato a columna Date**

```python
apple_data.loc[:, 'Date'] = pd.to_datetime(apple_data['Date'], format='%b %d, %Y')
apple_data.head()
```

**Paso 7: Verificar formato correcto de las columnas y valores nulos**

```python
apple_data.info()
```

**Paso 8 :Gráficar**

```python
apple_data = apple_data[apple_data["High"] >= 1]
apple_data["Open"].plot()
```

## Análisis y Conclusión

### Riesgo

**Cálculo del riesgo de Apple** 

```python
apple_data["risk"] =apple_data['Close'].pct_change()
apple_risk = apple_data['risk'].std() * 100
apple_risk
```

**Cálculo del riesgo de Microsoft** 

```python
microsoft_data["risk"] =microsoft_data['Close'].pct_change()
microsoft_risk = microsoft_data['risk'].std() * 100
microsoft_risk
```

**Cálculo del riesgo de Nvidia** 

```python
nvidia_data["risk"] =nvidia_data['Close'].pct_change()
nvidia_risk = nvidia_data['risk'].std() * 100
nvidia_risk
```

**Cálculo del riesgo de Alphabet** 

```python
Alphabet_data['risk'] = Alphabet_data['Close'].pct_change()
alphabet_risk = Alphabet_data['risk'].std() * 100
alphabet_risk
```

**Cálculo del riesgo de Meta** 

```python
meta_data['risk'] = meta_data['Close'].pct_change()
meta_risk = meta_data['risk'].std() * 100
meta_risk
```

## Gráficos

### Riesgo de las acciones

```python
risks = {
    'Meta': meta_risk,
    'Microsoft': microsoft_risk,
    'Apple': apple_risk,
    'Nvidia': nvidia_risk,
    'Alphabet': alphabet_risk
}

colors = ['blue', 'green', 'red', 'orange', 'purple']


plt.figure(figsize=(10, 6))
plt.bar(risks.keys(), risks.values(), color=colors)

plt.title('Riesgo de Acciones', fontsize=14)
plt.xlabel('Acciones', fontsize=12)
plt.ylabel('Volatilidad Diaria (%)', fontsize=12)

plt.show()
```

## Valor máximo de las acciones 

```python

meta_data['Date'] = pd.to_datetime(meta_data['Date'])
microsoft_data['Date'] = pd.to_datetime(microsoft_data['Date'])
apple_data['Date'] = pd.to_datetime(apple_data['Date'])
nvidia_data['Date'] = pd.to_datetime(nvidia_data['Date'])
Alphabet_data['Date'] = pd.to_datetime(Alphabet_data['Date'])

plt.figure(figsize=(12, 6))

plt.plot(meta_data['Date'], meta_data['High'], label='Meta', color='blue')
plt.plot(microsoft_data['Date'], microsoft_data['High'], label='Microsoft', color='green')
plt.plot(apple_data['Date'], apple_data['High'], label='Apple', color='red')
plt.plot(nvidia_data['Date'], nvidia_data['High'], label='Nvidia', color='orange')
plt.plot(Alphabet_data['Date'], Alphabet_data['High'], label='Alphabet', color='purple')

plt.title('Valor máximo de las acciones', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Valor Máximo (High)', fontsize=12)

plt.xticks(rotation=45)

plt.legend()

plt.tight_layout()
plt.show()
```

# Conclusión

## Análisis de Riesgo y Rentabilidad de Acciones

- **Nvidia** es la acción con mayor volatilidad (riesgo), lo que indica un mayor potencial de fluctuación en su precio. Además, es también la acción más económica en comparación con las demás.
  
- **Microsoft** presenta el menor riesgo (volatilidad) entre las acciones analizadas. A pesar de su estabilidad, es la segunda acción más cara después de Meta.
  
- En un **escenario adverso al riesgo**, es preferible optar por **Microsoft**, debido a su estabilidad y menor volatilidad.

- En un **esamantepropenso al riesgo**, **Alphabet** sería la opción más atractiva, dado su mayor potencial de rendimiento, aunque con más incertidumbre en  luego de **Nvidia**su valor.

- Si te encuentras en un **escenario neutral** con respecto al riesgo, **Apple** es la mejor opción, ya que ofrece un equilibrio entre volatilidad y rendimiento.
